```python
# Fonction pour extraire les informations
def extract_information(row):
    match = re.match(r'([A-Z\s]+)\s+(\d{2}-\d{2}-\d{4}|\d{4}-\d{2}-\d{2})\s+(\d{2}:\d{2})\s+(\d+)', row)
    if match:
        prenom = match.group(1).strip()
        date = match.group(2)
        heure = match.group(3)
        transaction_num = match.group(4)
        return pd.Series([prenom, date, heure, transaction_num])
    else:
        return pd.Series(['', '', '', ''])  # Retourner des valeurs vides si le format ne correspond pas

# Appliquer la fonction et créer de nouvelles colonnes
df_rua[['Prenom', 'Date', 'Heure', 'Transaction']] = df_rua['Unnamed: 2'].apply(extract_information)

```

```python
# Fonction pour extraire le prix
def extract_price(row):
    match = re.search(r'€(\d+\.\d{2})', row)
    if match:
        return match.group(1)
    else:
        return ''

# Appliquer la fonction et créer une nouvelle colonne 'Prix'
df_rua['Prix'] = df_rua['Unnamed: 2'].apply(extract_price)
```

```python
# Fonction pour extraire le total de la transaction
def extract_total_transaction(data):
    total_transaction = ''
    for cell in data.split('\n'):
        match = re.search(r'_T_O_T_A_L\s*€(\d+\.\d{2})', cell)
        if match:
            total_transaction = match.group(1)
    return total_transaction

# Appliquer la fonction et créer une nouvelle colonne 'Total Transaction'
df_rua['Total Transaction'] = df_rua['Unnamed: 2'].apply(extract_total_transaction)
```

```python
# Fonction pour extraire le total de la transaction
def extract_total_transaction(row):
    # Expression régulière pour trouver le total de la transaction avec underscores
    match = re.search(r'_T_O_T_A_L\s*_\s*_\s*_€_([0-9_]+(?:\.\d{2}))', row)
    if match:
        # Nettoyer le montant pour retirer les underscores
        total_str = match.group(1).replace('_', '')
        return total_str
    else:
        return ''

# Appliquer la fonction à chaque cellule de la colonne 'Unnamed: 2' et créer une nouvelle colonne 'Total Transaction'
df_rua['Total Transaction'] = df_rua['Unnamed: 2'].apply(extract_total_transaction)
```

```python
# Fonction pour extraire le total de la transaction
def extract_total_transaction(row):
    match = re.search(r'_T_O_T_A_L\s*_[\s\d_]*€([\d._]+)', row)
    if match:
        total_str = match.group(1).replace('_', '').replace('.', '').replace(',', '.')
        return total_str
    else:
        return ''

# Appliquer la fonction et créer une nouvelle colonne 'Total Transaction'
df_rua['Total Transaction'] = df_rua['Unnamed: 2'].apply(extract_total_transaction)/100
```
